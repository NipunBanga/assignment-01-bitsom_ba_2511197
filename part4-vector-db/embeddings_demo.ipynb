# Embeddings Demo

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import numpy as np
```

```python
sentences = [
# Cricket
"The batsman scored a century in the match",
"The bowler took three wickets in one over",
"The team won the cricket match convincingly",
# Cooking
"Cooking pasta requires boiling water and adding sauce",
"The chef prepared a delicious meal",
"Baking a cake requires flour and sugar",
# Cybersecurity
"Cybersecurity protects systems from attacks",
"Strong passwords help secure accounts",
"Phishing attacks trick users into revealing data",
"Encryption ensures secure communication"
]
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)
```

```python
similarity_matrix = cosine_similarity(embeddings)
plt.imshow(similarity_matrix)
plt.colorbar()
plt.title('Cosine Similarity Heatmap')
plt.show()
```

```python
query = "The bowler took three wickets in one over"
query_embedding = model.encode([query])
scores = cosine_similarity(query_embedding, embeddings)[0]

top_indices = scores.argsort()[-2:][::-1]

for idx in top_indices:
    print(sentences[idx], scores[idx])
```
